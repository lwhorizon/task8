```python
# 各用户购买数量分析
```

```python
#各用户累计购买数量分析
data1 = data.drop(['time'], axis=1) 
```

```python
userNum = data1.groupby('user', as_index=True).sum().drop(['money'], axis=1)
```

```python
userNum
```

```python
userNum.plot()
```

```python
#各用户购买数量累积贡献分析
userNumRate = userNum / 23570 * 100
```

```python
userNumRate.plot()
```

```python
plt.scatter(userNumRate.index,userNumRate)
```

```python
#各用户最大单笔购买数量分析
userOrderMax = data1.groupby('user', as_index=True).max()
```
