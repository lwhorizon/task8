```python
#各用户留存率
BuyFirstTime = data.groupby('user', as_index=True).min().drop(['money'], axis=1)

BuyFirstTime = BuyFirstTime.reset_index()

userRetention = pd.merge(left=data, right=BuyFirstTime, how='inner', on='user', suffixes = ('', '_min'))

userRetention = userRetention.drop(['num_min'], axis=1)

userRetention.head(10)
```

```python
userRetention['time'] = pd.to_datetime(userRetention['time'],format='%Y%m%d')
userRetention['time_min'] = pd.to_datetime(userRetention['time_min'],format='%Y%m%d')

userRetention['orderDateDiff'] = userRetention.time- userRetention.time_min

userRetention['dateDiff'] = userRetention.orderDateDiff.apply(lambda x: x/np.timedelta64(1, 'D'))

# userRetention['dateDiff'] = (userRetention['dateDiff'].astype(int))

userRetention
```

```python
#各用户首次回购周期
# userFirstReturnBuy = userRetention.groupby('user').min().dateDiff
userRetention['_time'] = pd.to_datetime(userRetention['time'], format='%Y%m%d').astype('datetime64[M]')
pivotedFirstReturnBuy = userRetention.pivot_table(index='user', columns='_time', 
                                                  values='dateDiff', aggfunc=min)

columnsMonth = userRetention._time.sort_values().astype('str').unique()
pivotedFirstReturnBuy.columns = columnsMonth

pivotedFirstReturnBuy1 = pivotedFirstReturnBuy.applymap(lambda x: np.NaN if x<=0 else x)

pivotedFirstReturnBuy1 = pivotedFirstReturnBuy1.min(axis=1)

pivotedFirstReturnBuy1.head()
```

```python
pivotedFirstReturnBuy1.hist(bins = 100,figsize=(12,6))
```

```python
timeBin = [0, 3, 7, 15, 30, 60, 90, 180, 365]
userRetention['dateDiffBin'] = pd.cut(userRetention.dateDiff, bins=timeBin)

userRetention
```

```python
pivotedRetention = userRetention.pivot_table(index='user', columns='dateDiffBin', 
                                             values='money', aggfunc=sum, dropna=False)
```

```python
pivotedRetention
```

```python
pivotedRetention.mean()
```

```python
pivotedRetentionTrans = pivotedRetention.fillna(0).applymap(lambda x: 1 if x>0 else 0)

pivotedRetentionTrans.head()
```

```python
pivotedRetentionTrans.mean().plot.bar()
```

```python
#各用户平均购买周期
def period(group):
    onePeriod = group.dateDiff - group.dateDiff.shift(-1)
    return onePeriod

allDiff = userRetention.groupby('user').apply(period)

allDiff.mean()
```

```python
allDiff.plot.hist(bins = 20)
```

```python
#各用户各月复购率
data['_time'] = pd.to_datetime(data['time'], format='%Y%m%d').astype('datetime64[M]')
pivotedCounts = data.pivot_table(index='user', columns='_time', values='time',aggfunc='count').fillna(0)

columnsMonth = data._time.sort_values().astype('str').unique()
pivotedCounts.columns = columnsMonth
pivotedCounts.head()
```

```python
pivotedCountsTransf = pivotedCounts.applymap(lambda x: 1 if x>1 else np.NaN if x==0 else 0)
pivotedCountsTransf.head()
```

```python
(pivotedCountsTransf.sum() / pivotedCountsTransf.count()).plot(figsize=(12,5))
```

```python
#各用户各月回购率
pivotedAmount = data.pivot_table(index='user', columns='_time', values='money', aggfunc='mean').fillna(0)

pivotedAmount.columns = columnsMonth
pivotedAmount.head()
```

```python
pivotedBuy = pivotedAmount.applymap(lambda x:1 if x>0 else 0)
pivotedBuy.head()
```

```python
def returnBuy(n):
    status = []
    for i in range(17):
        if n[i] == 1:
            if n[i+1] == 1:
                status.append(1)
            if n[i+1] == 0:
                status.append(0)
        else:
            status.append(np.NaN)
    status.append(np.NaN)
    return status
#每列进行比较，查看后面一列是否进行购买
pivotedReturnBuy = pivotedBuy.apply(returnBuy, axis=1)

pivotedReturnBuy1 = pd.DataFrame(list(pivotedReturnBuy), columns=columnsMonth,index=pivotedReturnBuy.index)

pivotedReturnBuy1.head()
```

```python
pivotedReturnBuy1.mean().plot(figsize=(12,5))
```

```python
#用户分层（活跃用户和回流用户比率）
def userStatus(n):
    status = []
    for i in range(18):
        #查看本月无消费
        if n[i] == 0:
            if len(status) > 0:
                if status[i-1] == 'unreg':
                    status.append('unreg')
                else:
                    status.append('unactive')
            else:
                status.append('unreg')
        #查看本月有消费
        else:
            if len(status) == 0:
                status.append('new')
            else:
                if status[i-1] == 'unactive':
                    status.append('return')
                elif status[i-1] == 'unreg':
                    status.append('new')
                else:
                    status.append('active')
    return status
pivotedBuyStatus = pivotedBuy.apply(lambda x: userStatus(x), axis=1)
pivotedBuyStatus1 = pd.DataFrame(list(pivotedBuyStatus), columns=columnsMonth,index=pivotedBuyStatus.index)

pivotedBuyStatus1.head()
```

```python
pivotedStatusCount = pivotedBuyStatus1.replace('unreg', np.NaN).apply(pd.value_counts)

pivotedStatusCount.head()
```

```python
# 蓝色部分代表郭跃用户
pivotedStatusCount.fillna(0).T.plot.area(figsize=(10,7))
```

```python
# 回流用户比率
userReturnRate = pivotedStatusCount.apply(lambda x: x/x.sum())

userReturnRate.loc['return'].plot(figsize = (12,6))
```

```python
# 活跃用户比率
userReturnRate.loc['active'].plot(figsize = (12,6))
```
